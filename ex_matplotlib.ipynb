```python
fig=plt.figure()
```

```python
axes1=fig.add_subplot(2,2,1)
axes2=fig.add_subplot(2,2,2)
axes3=fig.add_subplot(2,2,3)
axes4=fig.add_subplot(2,2,4)
```

```python
fig
```

```python
oneDF=anscombe[anscombe['dataset']=='I']
twoDF=anscombe[anscombe['dataset']=='II']
threeDF=anscombe[anscombe['dataset']=='III']
fourDF=anscombe[anscombe['dataset']=='IV']
```

```python
axes1.plot(oneDF.x,oneDF.y,'g+')
axes2.plot(twoDF.x,twoDF.y,'r.')
axes3.plot(threeDF.x,threeDF.y,'bo')
axes4.plot(fourDF.x,fourDF.y,'y^')
```

```python
axes1.set_title('I')
axes2.set_title('II')
axes3.set_title('III')
axes4.set_title('IV')
fig.suptitle('anscombe')
```
